### 2.2 Bivariate and Multivariate Analysis
**Bivariate** and **multivariate analysis** involve analyzing the relationships between two or more variables.
<br>We can use a **variety of plots** to visualize these relationships:
- **Scatter plots** (numerical vs. numerical)
- **Bar plots** (categorical vs. numerical)
- **Box plots and Violin plots** (categorical vs. numerical)
- **Displots** (numerical vs. categorical)
- **Pair plots** (multiple numerical variables)
- **Line plots** (numerical vs. numerical over time)

We can also use **heatmaps** to visualize the relationships between variables:
- **Contingency heatmaps** (categorical vs. categorical with cross‑tabulation counts)
- **Pivot table heatmaps** (categorical vs. categorical with aggregated values)
- **Correlation heatmaps** (correlation matrix of numerical variables)

A **correlation analysis** can be performed to quantify the strength and direction of relationships between variables, using appropriate correlation coefficients based on the types of variables being analyzed.

#### 2.2.1 Scatter Plots (Numerical vs. Numerical)
**Scatter plots** are used to visualize the relationship between two numerical variables.
<br>For example, we can create a scatter plot to analyze the relationship between `total_bill` and `tip` in the Tips dataset.

```python
# Create a scatter plot to analyze the relationship between total bill and tip
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

We can also improve the visualization by coloring the points based on a categorical variable, such as `gender`, and styling the points based on another categorical variable, such as `smoker` status or `size` of the party. This allows us to see how these additional variables may influence the relationship between `total_bill` and `tip`.

```python
# Create a scatter plot to analyze the relationship between total bill and tip, colored by gender
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Create a scatter plot to analyze the relationship between total bill and tip,
# colored by gender and styled by size of the party.
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', size='size')
plt.title('Scatter Plot of Total Bill vs Tip with Smoker Status and Party Size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Create a scatter plot to analyze the relationship between total bill and tip,
# colored by gender and styled by smoker status and size of the party.
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
plt.title('Scatter Plot of Total Bill vs Tip with Smoker Status and Party Size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

#### 2.2.3.1 Box Plots (Categorical vs. Numerical)
**Box plots** are another useful tool for visualizing the relationship between a categorical variable and a numerical variable.
<br>For example, we can create a box plot to analyze the distribution of `age` by `gender` and `survival status` in the Titanic dataset. This allows us to see how the age distribution varies between genders and between those who survived and those who did not survive.

```python
# Box plot to analyze the distribution of age by gender and survival status
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
plt.title('Box Plot of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

#### 2.2.3.2 Violin Plots (Categorical vs. Numerical)
**Violin plots** are similar to box plots but also show the kernel density estimation of the data, providing a more detailed view of the distribution of the numerical variable across different categories. For example, we can create a violin plot to analyze the distribution of `age` by `gender` and `survival status` in the Titanic dataset, similar to the box plot, but with additional information about the density of the age distribution within each category.

```python
# Violin plot to analyze the distribution of age by gender and survival status
sns.violinplot(data=titanic, x='sex', y='age', hue='survived')
plt.title('Violin Plot of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

Violin plots can also be split by a categorical variable, such as `survival status`, to compare the distributions of `age` for survivors and non-survivors within each gender.

```python
# Violin plot to analyze the distribution of age by gender, split by survival status
sns.violinplot(data=titanic, x='sex', y='age', hue='survived', split=True)
plt.title('Violin Plot of Age by Gender, split by Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

#### 2.2.4 Displots (Numerical vs. Categorical)
**Displots** are used to visualize the distribution of a numerical variable across different categories. For example, we can create a displot to analyze the distribution of `age` for `survivors` and `non-survivors` in the Titanic dataset, allowing us to see how the age distribution differs between those who survived and those who did not survive.

```python
# Distribution of age for non-survivors with a KDE curve
sns.displot(titanic[titanic['survived']==0]['age'], kde=True) 
plt.title('Distribution of Age for Non-Survivors with KDE')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of age for survivors and non-survivors with KDE curves
sns.displot(data=titanic, x='age', hue='survived', kind='hist', kde=True, element='step', common_norm=False)
plt.title('Age Distribution for Survivors vs Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

In a displot, the `count` option displays the number of observations in each bin of the histogram, while the `density` option normalizes the histogram so that the area under the curve sums to 1, representing the probability density function of the data. The `density` option allows for easier comparison between different groups or distributions, as it accounts for differences in sample size.

```python
# Plot the distribution of age for survivors and non-survivors with KDE curves,
# using a density plot instead of a histogram
sns.displot(data=titanic, x='age', hue='survived', kind='kde', fill=True, common_norm=False)
plt.title('Age Distribution for Survivors vs Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

#### 2.2.6 Line Plots (Numerical vs. Numerical over Time)
**Line plots** are used to visualize the relationship between two numerical variables, especially when one of the variables represents time.
<br>For example, we can create a line plot to analyze the trend of the number of passengers over the years in the flights dataset, allowing us to see how the number of passengers has changed over time.

#### 2.2.7 Contingency Heatmaps (Categorical vs. Categorical with Cross‑Tabulation Counts)
**Contingency heatmaps** are used to visualize the relationship between two categorical variables by displaying the counts of observations in each category combination.
<br>For example, we can create a contingency heatmap to analyze the relationship between passenger `class` and `survival status` in the Titanic dataset, allowing us to see how many passengers in each class survived or did not survive.

```python
# Create a contingency heatmap to analyze the relationship between passenger class and survival status
ax = sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']), annot=True, fmt='d', cmap='YlGnBu')
ax.figure.axes[-1].set_ylabel('Count')
plt.title('Contingency Heatmap of Passenger Class vs Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Passenger Class')
plt.show()
```

Now let's apply the contingency heatmap to analyze the relationship between `gender` and `survival status` in the Titanic dataset, allowing us to see how many passengers of each gender survived or did not survive.

```python
# Create a contingency heatmap to analyze the relationship between gender and survival status
ax = sns.heatmap(pd.crosstab(titanic['sex'], titanic['survived']), annot=True, fmt='d', cmap='YlGnBu')
ax.figure.axes[-1].set_ylabel('Count')
plt.title('Contingency Heatmap of Gender vs Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Gender')
plt.show()
```

#### 2.2.8 Pivot Table Heatmaps (Categorical vs. Categorical with Aggregated Values)
**Pivot table heatmaps** are used to visualize the relationship between two categorical variables by displaying aggregated values (such as counts, means, or sums) in a heatmap format.
<br>For example, we can create a pivot table heatmap to analyze the number of passengers by `month` and `year` in the flights dataset, allowing us to see how the number of passengers varies across different months and years.

```python
# Plot a pivot table heatmap to analyze the number of passengers by month and year in the flights dataset
pivot = flights.pivot_table(values='passengers', index='month', columns='year', observed=False, ).astype(int)
pivot
```

```python
# Plot the heatmap of the pivot table to visualize the number of passengers by month and year
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Pivot Table Heatmap of Number of Passengers by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

#### 2.2.9 Correlation Heatmaps (Numerical vs. Numerical with Correlation Matrix)
**Correlation heatmaps** are used to visualize the relationships between multiple numerical variables by displaying the correlation coefficients in a heatmap format.
<br>For example, we can create a correlation heatmap to analyze the relationships between the numerical variables in the datasets, allowing us to see how strongly each pair of variables is correlated with each other.

```python
# Plot a heatmap to visualize the correlation between numerical variables in the iris dataset
corr = iris.corr(numeric_only=True)
corr
```

The correlation coefficients can range from -1 to 1, where values close to 1 indicate a strong positive correlation, values close to -1 indicate a strong negative correlation, and values close to 0 indicate little to no correlation.

```python
# Plot the heatmap of the correlation matrix to visualize the relationships between numerical variables in the iris dataset
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap of Numerical Variables in the Iris Dataset')
plt.show()
```

```python
# Plot a correlation matrix heatmap to analyze the relationships between numerical variables in the titanic dataset
corr_titanic = titanic.corr(numeric_only=True)
sns.heatmap(corr_titanic, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap of Numerical Variables in the Titanic Dataset')
plt.show()
```

#### 2.2.10 Correlation Analysis

**Correlation analysis** involves calculating the correlation coefficients between pairs of numerical variables to quantify the strength and direction of their relationships.

##### Numerical Variables

We can calculate the **Pearson’s product moment correlation coefficients** between pairs of numerical variables in the dataset to understand how they are related to each other.

For example, we can calculate the Pearson’s product moment coefficient between `total_bill` and `tip` in the tips dataset to quantify the strength and direction of their relationship.

```python
# Calculate the Pearson’s product moment coefficient between total bill and tip in the tips dataset
pearson_corr = tips['total_bill'].corr(tips['tip'])
print(f"Pearson's correlation coefficient between total bill and tip: {pearson_corr:.2f}")

# Test the statistical significance of the correlation between total bill and tip using a t-test
from scipy.stats import pearsonr
corr, p_value = pearsonr(tips['total_bill'], tips['tip'])
print(f"P-value: {p_value:.4f}")

# Plot a scatter plot to visualize the relationship between total bill and tip
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'alpha':0.6})
plt.title('Scatter Plot of Total Bill vs Tip (R={:.2f})'.format(pearson_corr))
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

##### Ordinal Variables

When analyzing the relationship between two ordinal variables, we can use **Spearman’s rank correlation coefficient** to quantify the strength and direction of their monotonic relationship. This method is appropriate for ordinal data because it assesses how well the relationship between the two variables can be described using a monotonic function, without assuming a linear relationship.

For example, we can calculate the Spearman rank correlation coefficient between `year` and `passengers` in the flights dataset to understand how the number of passengers has changed over time.

```python
# Calculate the Spearman rank correlation coefficient between two ordinal variables in the flights dataset
spearman_corr = flights['year'].corr(flights['passengers'], method='spearman')
print(f"Spearman's rank correlation coefficient between passenger class and survival status: {spearman_corr:.2f}")

# Plot a scatter plot to visualize the relationship between passenger class and survival status
sns.regplot(data=flights, x='year', y='passengers', scatter_kws={'alpha':0.6})
plt.title('Scatter Plot of Year vs Passengers (Spearman R={:.2f})'.format(spearman_corr))
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.show()
```

##### Binary vs. Continuous Variables

When analyzing the relationship between two binary variables, we can use **Point-biserial correlation** to quantify the strength of association between them. This correlation coefficient is a special case of the Pearson correlation coefficient that is used when one variable is binary and the other variable is continuous.

For example, we can calculate the point-biserial correlation coefficient between `age` and `survival status` in the Titanic dataset to understand how age is associated with survival status.

```python
# Calculate the point-biserial correlation coefficient between age and survival status in the Titanic dataset
from scipy.stats import pointbiserialr

# Drop rows with missing values in the age and survived columns to ensure accurate correlation calculation
titanic_no_na = titanic[['age', 'survived']].dropna()

# Calculate the point-biserial correlation coefficient and p-value
corr, p_value = pointbiserialr(titanic_no_na['age'], titanic_no_na['survived'])
print(f"Point-biserial correlation coefficient between age and survival status: {corr:.2f}. P-value: {p_value:.2f}")

# Plot a scatter plot to visualize the relationship between age and survival status
sns.regplot(data=titanic_no_na, x='age', y='survived', scatter_kws={'alpha':0.6})
plt.title('Scatter Plot of Age vs Survival Status (Point-Biserial R={:.2f})'.format(corr))
plt.xlabel('Age')
plt.ylabel('Survival Status (0=Deceased, 1=Survived)')
plt.show()
```

##### Categorical Variables

When analyzing the relationship between two categorical variables, we can use **Chi-square test of independence** to determine if there is a significant association between the two variables. This test assesses whether the observed frequencies in a contingency table differ significantly from the expected frequencies under the assumption of independence.
<br>In other words, it tests the hypothesis that variables are independent of one another.

For example, we can compute the chi-square test of independence between passenger `class` and `survival status` in the Titanic dataset to determine if there is a significant association between these two categorical variables.

```python
# Compute the chi-square test of independence between passenger class and survival status in the Titanic dataset
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(titanic['pclass'], titanic['survived'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2:.2f}, P-value: {p:.4f}, Degrees of freedom: {dof}")
```

> A large χ² value means the observed counts differ a lot from what independence would predict → strong evidence of association.

> A small χ² value means the observed counts are close to expected → likely no association.
